```python
class SparseMatrixCOO:
    def __init__(self, shape):
        self.shape = shape
        self.data = []
        self.row = []
        self.col = []

    def add_value(self, row, col, value):
        if value != 0:
            self.row.append(row)
            self.col.append(col)
            self.data.append(value)

    def to_dense(self):
        dense_matrix = [[0] * self.shape[1] for _ in range(self.shape[0])]
        for r, c, v in zip(self.row, self.col, self.data):
            dense_matrix[r][c] = v
        return dense_matrix

    def __str__(self):
        return f"Data: {self.data}\nRow indices: {self.row}\nColumn indices: {self.col}"

# Ejemplo de uso
sparse_matrix = SparseMatrixCOO((4, 4))
sparse_matrix.add_value(0, 0, 1)
sparse_matrix.add_value(1, 2, 3)
sparse_matrix.add_value(3, 1, 4)

print("Sparse matrix (COO format):")
print(sparse_matrix)
print("\nDense representation:")
for row in sparse_matrix.to_dense():
    print(row)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def taylor_series(f, a, n):
    x = sp.symbols('x')
    series = 0
    for i in range(n+1):  # Incluye el término de orden 0
        term = (f.diff(x, i).subs(x, a) * (x - a)**i) / sp.factorial(i)
        series += term
    return series

def get_function_choice():
    print("Elige una función:")
    print("1. sin(x)")
    print("2. cos(x)")
    print("3. exp(x)")
    print("4. log(x)")
    print("5. x**2")
    choice = int(input("Ingresa el número de la función deseada: "))
    return choice

def main():
    # Elegir función
    choice = get_function_choice()
    x = sp.symbols('x')
    if choice == 1:
        f = sp.sin(x)
    elif choice == 2:
        f = sp.cos(x)
    elif choice == 3:
        f = sp.exp(x)
    elif choice == 4:
        f = sp.log(x)
    elif choice == 5:
        f = x**2
    else:
        raise ValueError("Opción no válida.")

    # Entrada del usuario
    a = float(input("Ingresa el punto de expansión: "))
    n = int(input("Ingresa la cantidad de términos de la expansión: "))

    # Calcular la serie de Taylor
    ye = taylor_series(f, a, n)
    print("Serie de Taylor:")
    print(ye)

    # Convertir la serie de Taylor en una función numérica
    ye_func = sp.lambdify(x, ye, 'numpy')

    # Rango para la gráfica
    x_vals = np.linspace(a - 2, a + 2, 400)  # Centrado alrededor del punto de expansión
    fx = np.array([f.evalf(subs={x: xi}) for xi in x_vals])
    p1 = ye_func(x_vals)

    # Graficar
    plt.plot(x_vals, fx, '-b', lw=2, label='Función original')
    plt.plot(x_vals, p1, '-k', lw=2, label=f'Serie de Taylor (orden {n})')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Comparación entre la función y la serie de Taylor')
    plt.grid(True)
    plt.show()


main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

Descenso de gradiente

```python
def objective_function(x):
    return x[0]**2 + x[1]**2

# Gradiente de la función objetivo
def gradient(x):
    return np.array([2*x[0], 2*x[1]])

# Hessiana de la función objetivo
def hessian(x):
    return np.array([[2, 0], [0, 2]])

def gradient_descent(starting_point, learning_rate, max_iter):
    x = np.array(starting_point)
    history = [x.copy()]
    for _ in range(max_iter):
        grad = gradient(x)
        x = x - learning_rate * grad
        history.append(x.copy())
        if np.linalg.norm(grad) < 1e-6:
            break
    return np.array(history)

```

newtow

```python
def newton_method(starting_point, max_iter):
    x = np.array(starting_point)
    history = [x.copy()]
    for _ in range(max_iter):
        grad = gradient(x)
        H = hessian(x)
        if np.linalg.det(H) == 0:  # Regularización si es necesario
            H += np.eye(len(x)) * 1e-5
        x = x - np.linalg.inv(H) @ grad
        history.append(x.copy())
        if np.linalg.norm(grad) < 1e-6:
            break
    return np.array(history)
```

BFGS

```python
from scipy.optimize import minimize
def bfgs_method(starting_point):
    result = minimize(objective_function, starting_point, method='BFGS', jac=gradient)
    return result.x
```

```python
starting_points = [(1, 2), (10, -10), (-5, 5)]
learning_rates = [0.1, 0.01]
max_iter = 100

for sp in starting_points:
    for lr in learning_rates:
        gd_history = gradient_descent(sp, lr, max_iter)
        newton_history = newton_method(sp, max_iter)
        bfgs_result = bfgs_method(sp)

        # Graficar resultados
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.plot(gd_history[:, 0], gd_history[:, 1], marker='o')
        plt.title(f'GD with lr={lr}, starting point={sp}')

        plt.subplot(1, 3, 2)
        plt.plot(newton_history[:, 0], newton_history[:, 1], marker='o')
        plt.title(f'Newton, starting point={sp}')

        plt.subplot(1, 3, 3)
        plt.plot(bfgs_result[0], bfgs_result[1], 'ro')
        plt.title(f'BFGS, starting point={sp}')

        plt.show()
```
